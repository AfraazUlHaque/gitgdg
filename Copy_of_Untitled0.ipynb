```python
import numpy as np

# Function to convert size string to MB
def convert_size_to_mb(size):
    if isinstance(size, str):
        if 'M' in size:
            return float(size.replace('M', ''))
        elif 'k' in size:
            return float(size.replace('k', '')) / 1024
        elif 'Varies with device' in size:
            return np.nan
    return np.nan # Default for any other unexpected format or non-string types

# Apply the function to the 'Size' column to create 'size_mb'
df['size_mb'] = df['Size'].apply(convert_size_to_mb)
```
